**Theoretical Questions**

**Question. No. 1. What is unsupervised learning in the context of machine learning?**

**Anwer.** Unsupervised learning is a machine learning approach where models are trained on unlabeled data without predefined outputs. The algorithm explores data structure, patterns, or relationships by itself. It is mainly used for clustering, dimensionality reduction, and association rule mining. Unlike supervised learning, it does not rely on labeled examples but identifies hidden patterns, similarities, or groupings. Common techniques include k-means clustering, hierarchical clustering, and principal component analysis (PCA). It helps in data exploration, pattern discovery, and anomaly detection.


**Question. No. 2. How does K-Means clustering algorithm work?**

**Answer.** K-Means clustering is an unsupervised learning algorithm that groups data into k clusters based on similarity. It works by initializing k centroids randomly, then assigning each data point to the nearest centroid. After assignment, the algorithm recalculates the centroids as the mean of all points in a cluster. This process of assignment and centroid update repeats until centroids stabilize or changes become minimal. The goal is to minimize within-cluster variance, ensuring data points in the same cluster are highly similar.

**Question. No. 3. Explain the concept of a dendrogram in hierarchical clustering.**

**Answer.** A dendrogram is a tree-like diagram used in hierarchical clustering to illustrate how data points are grouped together at various similarity levels. It starts with each data point as an individual cluster and then progressively merges the closest clusters based on a distance metric. The vertical axis represents the distance or dissimilarity, while the horizontal axis shows data points or clusters. By cutting the dendrogram at a chosen height, specific clusters can be formed. It visually explains the hierarchical structure of clusters.


**Question. No. 4. What is the main difference between K-Means and Hierarchical Clustering?**

** Answer.** The main difference between K-Means and Hierarchical Clustering lies in their approach and output. K-Means is a partitioning method that requires specifying the number of clusters (k) in advance and groups data by minimizing within-cluster variance. It works best for large datasets but may give different results on each run. Hierarchical Clustering, on the other hand, builds a tree-like structure (dendrogram) without predefining clusters. Clusters are formed by cutting the dendrogram, making it more interpretable but less efficient for very large datasets.


**Question. No. 5. What are the advantages of DBSCAN over K-Means?**

**Answer.** DBSCAN offers several advantages over K-Means in clustering tasks. Unlike K-Means, it does not require specifying the number of clusters in advance. It can find clusters of arbitrary shape, whereas K-Means mainly works well with spherical clusters. DBSCAN also identifies noise and outliers effectively, labeling them separately instead of forcing them into clusters. Additionally, it handles datasets with clusters of varying densities better than K-Means, which often fails in such situations. This makes DBSCAN more flexible for real-world data analysis.


**Question. No. 6. When would you use Silhouette Score in clustering?**

**Answer.** The Silhouette Score is used in clustering to measure how well data points fit within their assigned clusters compared to other clusters. It evaluates both cohesion (how close a point is to its own cluster) and separation (how far it is from other clusters). The score ranges from -1 to +1, where values close to +1 indicate well-separated clusters, near 0 suggest overlapping clusters, and negative values imply misclassification. It is commonly used to choose the optimal number of clusters and assess clustering quality.

** Question. NO. 7. What are the limitations of Hierarchical Clustering?**

**Answer.**Hierarchical Clustering has several limitations despite its interpretability. First, it is computationally expensive, especially with large datasets, since it calculates pairwise distances repeatedly. Second, once a merge or split is made, it cannot be undone, which may lead to suboptimal clustering. Third, it is sensitive to noise and outliers, which can distort the dendrogram. Additionally, the choice of linkage method (single, complete, average) can significantly affect results. Lastly, it struggles with very large datasets compared to faster algorithms like K-Means.

**Question. No. 8. Why is feature scaling important in clustering algorithms like K-Means?**

**Answer.** Feature scaling is important in clustering algorithms like K-Means because these methods rely on distance calculations (such as Euclidean distance) to assign points to clusters. If features are on different scales, variables with larger ranges may dominate the distance measure, leading to biased clustering results. Scaling ensures that each feature contributes equally to the clustering process. Techniques like standardization (z-score normalization) or min-max scaling are commonly used, making clustering more accurate, fair, and reflective of true data patterns.


**Question. No. 9. How does DBSCAN identify noise points?**

**Answer.** DBSCAN identifies noise points based on density criteria. It classifies data points into three categories: core points, border points, and noise (outliers). A point is a core point if it has at least a minimum number of neighboring points (MinPts) within a given distance (ε). Border points are close to core points but have fewer neighbors than MinPts. Any point that is neither a core nor a border point is labeled as noise, meaning it does not belong to any cluster.


**Question. No. 10. Define inertia in the context of K-Means.**

**Answer.** Inertia in the context of K-Means refers to the sum of squared distances between each data point and its assigned cluster centroid. It measures how tightly the data points are grouped within clusters. A lower inertia value indicates more compact and well-formed clusters, while a higher value suggests that points are spread out and clusters may not be well-defined. Inertia is commonly used as a performance metric in K-Means and helps in determining the optimal number of clusters using the Elbow Method.


**Question. No.11. What is the elbow method in K-Means clustering?**

**Answer.** The Elbow Method is a technique used in K-Means clustering to determine the optimal number of clusters. It involves plotting the inertia (sum of squared distances between points and their centroids) against different values of k (number of clusters). As k increases, inertia decreases, but the rate of improvement slows. The point on the curve where the decrease sharply shifts to a gradual decline forms an “elbow.” This elbow indicates the best balance between cluster compactness and simplicity, suggesting the ideal number of clusters.


**Question. No. 12. Describe the concept of "density" in DBSCAN.**

**Answer.** In DBSCAN, density refers to how closely data points are packed together within a region. The algorithm defines density using two parameters: ε (epsilon), which is the maximum distance to consider neighboring points, and MinPts, the minimum number of points required to form a dense region. A point is considered a core point if it has at least MinPts within ε distance. Dense areas form clusters, while sparse regions with low density are treated as noise or outliers.


**Question. No. 13. Can hierarchical clustering be used on categorical data?**

**Answer.** Yes, hierarchical clustering can be used on categorical data, but it requires appropriate distance or similarity measures since traditional Euclidean distance is not suitable. For categorical variables, measures like Hamming distance, Jaccard similarity, or simple matching coefficient are often applied. Specialized methods such as hierarchical clustering with Gower’s distance can also handle mixed data types (numerical + categorical). However, results may be less interpretable compared to numerical data, and preprocessing or encoding techniques may be necessary for better clustering performance.

**Question. No. 14. What does a negative Silhouette Score indicate?**

**Answer.** A negative Silhouette Score indicates that the clustering is poor and some data points may have been assigned to the wrong cluster. The score compares how similar a point is to its own cluster versus the nearest neighboring cluster. When the score is negative, it means the point is closer to another cluster than the one it belongs to. This suggests overlapping or poorly separated clusters, and the chosen clustering parameters, such as the number of clusters, may need to be re-evaluated.

**Question. No. 15. Explain the term "linkage criteria" in hierarchical clustering.**

**Answer.** In hierarchical clustering, the linkage criteria determines how the distance between two clusters is calculated when merging them. It plays a crucial role in shaping the cluster structure. Common linkage methods include single linkage, which considers the minimum distance between cluster points, complete linkage, which uses the maximum distance, and average linkage, which computes the mean distance between all points of the clusters. Another method, Ward’s linkage, focuses on minimizing the variance within clusters. The selected linkage criterion directly impacts the dendrogram and clustering results.

**Question. No. 16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?**

**Answer.** K-Means clustering may perform poorly on data with varying cluster sizes or densities because it assumes clusters are spherical, equally sized, and evenly dense. The algorithm minimizes variance around centroids, so larger or denser clusters dominate, pulling centroids toward them. Smaller or less dense clusters may get merged with nearby larger ones, leading to misclassification. Additionally, K-Means relies on Euclidean distance, which does not adapt well to clusters of irregular shapes or different densities. This limitation makes it unsuitable for highly imbalanced cluster distributions.

**Question. No. 17. What are the core parameters in DBSCAN, and how do they influence clustering?**

**Answer.** The core parameters in DBSCAN are Epsilon (ε) and MinPts (Minimum Points), which directly influence clustering results. Epsilon (ε) defines the maximum radius around a point to search for neighbors. A small ε may lead to many points being labeled as noise, while a large ε may merge distinct clusters. MinPts specifies the minimum number of points required to form a dense region. Higher MinPts increases the strictness for defining clusters, while lower MinPts may allow smaller clusters. Together, these parameters control cluster shape, density, and noise handling.

**Question. No. 18. How does K-Means++ improve upon standard K-Means initialization?**

**Answer.** K-Means++ improves upon standard K-Means initialization by selecting initial centroids in a smarter way to enhance clustering performance. In standard K-Means, centroids are chosen randomly, which can lead to poor convergence, empty clusters, or results stuck in local minima. K-Means++ starts by selecting the first centroid randomly, then chooses subsequent centroids with a probability proportional to their squared distance from the nearest existing centroid. This ensures centroids are well spread out. As a result, K-Means++ provides faster convergence, better accuracy, and more stable clustering outcomes.

**Question. No. 19. What is agglomerative clustering?**

**Answer.** Agglomerative clustering is a type of hierarchical clustering that builds clusters in a bottom-up approach. Initially, each data point is treated as an individual cluster. The algorithm then repeatedly merges the two closest clusters based on a chosen distance metric and linkage criteria until all points belong to a single cluster or a stopping condition is met. The results are represented in a dendrogram, which shows the hierarchy of cluster merging. Agglomerative clustering is widely used because it does not require predefining the number of clusters.

**Question. No. 20. What makes Silhouette Score a better metric than just inertia for model evaluation?**

**Answer.** The Silhouette Score is considered a better metric than just inertia for evaluating clustering models because it balances both cohesion (how close points are within a cluster) and separation (how far clusters are from each other). Inertia only measures within-cluster variance and always decreases as the number of clusters increases, which can be misleading. In contrast, the Silhouette Score ranges from -1 to +1, providing insights into the quality of cluster assignments. A higher score indicates well-defined clusters, making it more reliable for model evaluation.

**Question. No. 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.**

**Answer.**

```python

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Visualization
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, alpha=0.75, marker='X', label="Centroids")
plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

**Question. No. 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.**

**Answer.**

```python

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

#Evaluate silhouette scores for k = 2 to 5
scores = []
k_values = [2, 3, 4, 5]

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

#Plot bar chart
plt.figure(figsize=(8,6))
plt.bar(k_values, scores, color="skyblue", edgecolor="black")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5", fontsize=14)
plt.show()
```

**Question. No. 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.**

**Answer.**

```python

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

#Generate synthetic data with overlapping clusters
X, y = make_blobs(n_samples=500, centers=3, cluster_std=2.0, random_state=42)

#Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

#Plot decision boundaries
plt.figure(figsize=(8,6))

# Create a mesh grid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))

# Predict cluster for each point in the grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries
plt.contourf(xx, yy, Z, alpha=0.3, cmap="viridis")

# Plot clustered points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=30, edgecolor="k")

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c="red", marker="X", s=200, label="Centers")

plt.title("K-Means Clustering with Decision Boundaries", fontsize=14)
plt.legend()
plt.show()
```

**Question. No. 34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.**

**Answer.**

```python

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

#Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data

#Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Compute inertia for K = 2 to 6
k_values = range(2, 7)
inertias = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

#Plot inertia values
plt.figure(figsize=(8,6))
plt.plot(k_values, inertias, marker="o", linestyle="-", color="b")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("K-Means Inertia for K = 2 to 6 (Breast Cancer Dataset)", fontsize=14)
plt.grid(True)
plt.show()
```

**Question. No. 44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.**

**Answer.**

```python

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np

#Load Iris dataset
iris = load_iris()
X = iris.data

#Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

#Compute Silhouette Coefficients
silhouette_vals = silhouette_samples(X, labels)
silhouette_avg = silhouette_score(X, labels)

#Visualization of Silhouette Coefficient for each sample
y_lower = 10
fig, ax = plt.subplots(figsize=(8, 6))

for i in range(3):  # for each cluster
    cluster_silhouette_vals = silhouette_vals[labels == i]
    cluster_silhouette_vals.sort()

    y_upper = y_lower + len(cluster_silhouette_vals)
    ax.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        cluster_silhouette_vals,
        alpha=0.7
    )
    ax.text(-0.05, y_lower + 0.5 * len(cluster_silhouette_vals), str(i))
    y_lower = y_upper + 10  # spacing between clusters

ax.axvline(silhouette_avg, color="red", linestyle="--")
ax.set_xlabel("Silhouette Coefficient Values")
ax.set_ylabel("Cluster")
ax.set_title("Silhouette Plot for KMeans Clustering (Iris Dataset)")
plt.show()
```

**Question. No. 46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pair plot (first 4 features).**

**Question. No. 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.**

**Answer.**

```python

import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

#Generate noisy blob data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Add random noise points
rng = np.random.RandomState(42)
noise = rng.uniform(low=-10, high=10, size=(50, 2))
X = np.vstack([X, noise])

#Apply DBSCAN
db = DBSCAN(eps=0.8, min_samples=5).fit(X)
labels = db.labels_

#Count clusters (excluding noise) and noise points
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")
```

**Question. No. 48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.**

**Answer.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

#Load Digits dataset
digits = load_digits()
X = digits.data

#Reduce dimensions using t-SNE (2D for visualization)
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_2d = tsne.fit_transform(X)

#Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_2d)

#Plot results
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_2d[:,0], X_2d[:,1], c=labels, cmap='tab10', s=10)
plt.title("Digits Dataset Clustering with t-SNE + Agglomerative Clustering")
plt.colorbar(scatter, label="Cluster")
plt.show()
```
